```python
input('Hint : ')
from getpass import getpass
string = getpass("Call Player 1\n Provide Puzzle Answer : ")
puzzle = string.lower()
if len(string) < 6 :
    print('Answer should be > 6 character with no spaces')
else:
    random_value,devid = '', ''
    false_attempt = 0
    #Hint if Any
    import math, random
    puzzle_type = input('Do you want system designed puzzle? y/n or Y/N : ')
    if puzzle_type == 'y' or puzzle_type == 'Y':
        for i in range((len(puzzle)//2)) :
            random_value = puzzle[math.floor(random.random()*len(puzzle))]
            if random_value not in devid:
                devid += random_value
        for a in range(len(puzzle)):
            if puzzle[a] in devid:
                puzzle = puzzle.replace(puzzle[a],'_')
    else:
        print('For Player 1 Only')
        num_blanks = int(input('How many blanks you prefer? : '))
        while len(devid) < num_blanks:
            blank_char = input('which character you want to blanked : ')
            if blank_char in puzzle:
                puzzle = puzzle.replace(blank_char,'_')
                devid += blank_char
                print(puzzle)
            else:
                print('Element not exist in Puzzle')
        
    print('\nCall Player 2\n Your Puzzle :',puzzle)
    while false_attempt < len(devid):
        if puzzle == string.lower():
                break
        ask = 'Remain Attempt : {} | Guess a Word (unique) : '.format(len(devid)-(false_attempt))
        guess = input(ask)
        if guess in devid and len(guess) == 1:
            for a in range(len(puzzle)):
                if string[a].lower() == guess:
                    puzzle = puzzle[:a] + guess + puzzle[a+1:]
            print('Keep it on ',puzzle)
        else:
            false_attempt += 1
            print('Oops! Try again')
    if puzzle == string.lower():
        print('\nCongratulation, Player 2 Won')
    else:
        print('\nCongratulation, Player 1 Won')
```
